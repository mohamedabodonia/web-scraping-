import libraries


```python
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlopen
```

inputing the url

```python
url ="https://wuzzuf.net/search/jobs/?q=illustrator"
```

create a connection to the url

```python
client = urlopen(url)
```

Getting the url code

```python
html =client.read()
```

```python
html

```

closing the request


```python
client.close()
```

creating an html parsing using beautifulsoup

```python
soup =bs(html,"html.parser")
soup
```

creating a container for all the needed

```python
container =soup("div",{"class":"css-1gatmva e1v1l3u10"})
len(container)
bs.prettify(container[0])
```

Accessing page item "jop title "

```python
joptitle = container[0].findAll("h2",{"class":"css-m604qf"})
joptitle[0].text
```

```python
cname = container[0].findAll("a",{"class":"css-17s97q8"})
cname[0].text.strip("-")
```

```python
joptype = container[0].findAll("span",{"class":"css-1ve4b75 eoyjyou0"})
joptype[0].text
```

Bringing Data

```python

file=open("/content/web scrabing.csv","w")
header ="Joptitle,Companyname,Joptype \n"
file.write(header)
```

```python
for c in container :
  joptitle = c.findAll("h2",{"class":"css-m604qf"})
  a=joptitle[0].text.strip()
  #print(a)
  cname = c.findAll("a",{"class":"css-17s97q8"})
  b=cname[0].text.strip("-")
  #print(b)

  joptype = c.findAll("span",{"class":"css-1ve4b75 eoyjyou0"})
  c=joptype[0].text
  #print(c)
  #print(a+","+b+","+c+",")
  file.write(a+","+b+","+c+ "\n")

file.close()
```
